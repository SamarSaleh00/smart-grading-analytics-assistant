### Step 2.1: Establish the Session Memory Buffer

This cell creates our state management system. It maintains the conversational history and ensures we do not exceed the token limits of our rewriting model by strictly keeping only the last 3 turns.

```python
# Cell 4: Dialogue State Tracker (DST)
class DialogueManager:
    def __init__(self, max_history=3):
        self.dialogue_history = []
        self.max_history = max_history

    def add_turn(self, role, text):
        """
        Appends a new turn to the history and trims older turns
        if we exceed the maximum history limit.
        """
        self.dialogue_history.append({"role": role, "text": text})
        
        # Keep only the last N pairs of interaction (User + System = 1 turn)
        # We multiply by 2 because each turn typically has a user and system response.
        if len(self.dialogue_history) > self.max_history * 2:
            self.dialogue_history = self.dialogue_history[-(self.max_history * 2):]

    def get_history_string(self):
        """
        Formats the history into a string for the LLM prompt.
        """
        history_str = ""
        for entry in self.dialogue_history:
            prefix = "User" if entry["role"] == "user" else "Assistant"
            history_str += f"{prefix}: {entry['text']}\n"
        return history_str.strip()

    def clear_history(self):
        self.dialogue_history = []
        print("🧹 System Notice: Dialogue history cleared.")

# Initialize our global session manager
session_manager = DialogueManager(max_history=3)
print("✅ System Notice: DialogueManager initialized.")
```

```python
from transformers import pipeline

print("⏳ System Notice: Loading FLAN-T5-Small model for query rewriting...")

print("✅ System Notice: Rewriter model loaded successfully.")

    """
    Takes the current user query and conversational history,
    and outputs a de-contextualized, explicit query.
    """
    history = dialogue_manager.get_history_string()
    
    # If there is no history, no rewriting is needed
    if not history:
        return current_query

    # Prompt engineered for FLAN-T5 to perform coreference resolution
    prompt = (
        "Given the following conversation history, rewrite the user's latest query to be "
        "fully independent and explicit. Replace pronouns like 'they', 'their', 'them', or 'it' "
        "with the exact subject they refer to from the history.\n\n"
        f"History:\n{history}\n\n"
        f"Latest User Query: {current_query}\n\n"
        "Rewritten Query:"
    )

    # Generate the rewritten query
    result = rewriter_pipeline(
        prompt,
        max_length=50,
        do_sample=False # Greedy decoding for more deterministic outputs
    )
    
    rewritten_text = result[0]['generated_text'].strip()
    return rewritten_text
```

````python
# Cell 4: Dialogue State Tracker (DST)
class DialogueManager:
    def __init__(self, max_history=3):
        self.dialogue_history = []
        self.max_history = max_history

    def add_turn(self, role, text):
        """
        Appends a new turn to the history and trims older turns
        if we exceed the maximum history limit.
        """
        self.dialogue_history.append({"role": role, "text": text})

        # Keep only the last N pairs of interaction (User + System = 1 turn)
        # We multiply by 2 because each turn typically has a user and system response.
        if len(self.dialogue_history) > self.max_history * 2:
            self.dialogue_history = self.dialogue_history[-(self.max_history * 2):]

    def get_history_string(self):
        """
        Formats the history into a string for the LLM prompt.
        """
        history_str = ""
        for entry in self.dialogue_history:
            prefix = "User" if entry["role"] == "user" else "Assistant"
            history_str += f"{prefix}: {entry['text']}\n"
        return history_str.strip()

    def clear_history(self):
        self.dialogue_history = []
        print("🧹 System Notice: Dialogue history cleared.")

# Initialize our global session manager
session_manager = DialogueManager(max_history=3)
print("✅ System Notice: DialogueManager initialized.")
````

````python
# Cell 5: Query Rewriting Module (Updated)
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

print("⏳ System Notice: Loading FLAN-T5-Small model for query rewriting...")

# Load model and tokenizer explicitly to avoid pipeline task name conflicts
model_id = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_id)
rewriter_model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

print("✅ System Notice: Rewriter model loaded successfully.")

def rewrite_query(current_query, dialogue_manager):
    """
    Takes the current user query and conversational history,
    and outputs a de-contextualized, explicit query.
    """
    history = dialogue_manager.get_history_string()

    # If there is no history, no rewriting is needed
    if not history:
        return current_query

    # Prompt engineered for FLAN-T5 to perform coreference resolution
    prompt = (
        "Given the following conversation history, rewrite the user's latest query to be "
        "fully independent and explicit. Replace pronouns like 'they', 'their', 'them', or 'it' "
        "with the exact subject they refer to from the history.\n\n"
        f"History:\n{history}\n\n"
        f"Latest User Query: {current_query}\n\n"
        "Rewritten Query:"
    )

    # Tokenize the input prompt
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate the rewritten query
    outputs = rewriter_model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=False # Greedy decoding for deterministic outputs
    )

    # Decode the output back into a readable string
    rewritten_text = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    return rewritten_text
````

### Step 3.2 & 3.3: Stateful Action Sequence & Execution Wrapper

This cell defines the retrieval engine. It converts numerical values to strings (as TAPAS requires), handles queries against either the whole dataset or the active subset, and maps TAPAS coordinate outputs back to actual Pandas rows to cache our `current_retrieved_subset`.

```python
# Cell 8: Table-QA Execution & Subset Caching Engine

# Global variable to cache the current active subset of students being discussed
current_retrieved_subset = None

def execute_table_qa(query, original_df):
    """
    Executes a query over the student dataset using TAPAS.
    Implements stateful routing: if a subset is cached and the query is a follow-up,
    it executes over the subset to maintain conversational context.
    """
    global current_retrieved_subset
    
    # Context Routing Logic: Decide whether to query the full table or the active subset
    # If the user is asking a follow-up ("their", "pattern", "average"), use the subset.
    is_follow_up = any(word in query.lower() for word in ["their", "them", "they", "average", "pattern", "this"])
    
    if is_follow_up and current_retrieved_subset is not None and not current_retrieved_subset.empty:
        active_df = current_retrieved_subset
        print("🔍 Context Router: Executing query against the cached subset of students.")
    else:
        active_df = original_df
        print("🔍 Context Router: Executing query against the entire student database.")

    # TAPAS requires all cells and column headers to be strings
    tapas_table = active_df.astype(str)
    
    try:
        # Run the TAPAS pipeline
        result = table_qa_pipeline(table=tapas_table, query=query)
        
        # Parse coordinates to isolate which rows the model selected
        coordinates = result.get("coordinates", [])
        cells = result.get("cells", [])
        aggregator = result.get("aggregator", "NONE")
        answer = result.get("answer", "")
        
        # If TAPAS targeted specific cells, let's extract their corresponding rows
        if coordinates:
            # Extract unique row indices chosen by TAPAS
            selected_row_indices = list(set([coord[0] for coord in coordinates]))
            
            # If we were querying the full table and looking for a list, update the global subset
            if not is_follow_up:
                current_retrieved_subset = active_df.iloc[selected_row_indices].copy()
        
        return {
            "answer": answer,
            "cells": cells,
            "aggregator": aggregator,
            "extracted_subset_size": len(current_retrieved_subset) if current_retrieved_subset is not None else 0
        }
        
    except Exception as e:
        print(f"❌ Table-QA Engine Error: {str(e)}")
        return {"answer": "Error processing table query.", "cells": [], "aggregator": "NONE", "extracted_subset_size": 0}

```

## 🧭 Phase 4 Implementation: Guardrails & Proactivity

### Step 4.1 & 4.2: Code the Clarification Gate & Proactive Pattern Detector

Run this cell to implement the business logic rules.

* The **Clarification Gate** checks for ambiguous column intents (e.g., saying "quiz" instead of specifying Quiz 1 or Quiz 2).
* The **Proactive Pattern Detector** checks if the active student cohort has an abnormally high rate of missing work and prepares a recommended action.

```python
# Cell 10: Dialogue Policy & Clarification Module

def check_clarification_gate(query):
    """
    Rule-based intercept to check if the user query is too ambiguous
    for the database columns. Returns a clarification string if ambiguous,
    otherwise returns None.
    """
    lowered_query = query.lower()
    
    # Check for ambiguous Quiz references
    if "quiz" in lowered_query and not any(x in lowered_query for x in ["1", "2", "one", "two"]):
```
